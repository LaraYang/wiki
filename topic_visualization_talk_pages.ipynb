## Visualization for UserTalk

```python
def generate_visualization(fn, num_topic, passes):    
    LDAvis_fn = os.path.join(model_dir, 'results', 'ldavis_' + fn + 'topic' + str(num_topic) + '_pass' + str(passes) + '.html')
    model = gensim.models.LdaModel.load(os.path.join(model_dir, fn + 'model.gensim'))
    corpus = pickle.load(open(os.path.join(model_dir, fn + 'corpus.pkl'), 'rb'))
    dictionary = corpora.Dictionary.load(os.path.join(model_dir, fn + 'dictionary.gensim'))
    LDAvis = pyLDAvis.gensim.prepare(model, corpus, dictionary)
    pyLDAvis.save_html(LDAvis, LDAvis_fn)
    pprint(model.print_topics())
    return LDAvis
```

```python
%%notify
usertalk_human_LDAvis = generate_visualization('usertalk_by_conv_human_', 15, 10)
```
