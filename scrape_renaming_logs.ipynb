```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from datetime import datetime
import os
```

```python
global_link = "https://meta.wikimedia.org/w/index.php?title=Special:Log/gblrename&type=gblrename&user="
local_link = "https://en.wikipedia.org/wiki/Special:Log/renameuser"
```

## Parsing Global Renaming Log

```python
next_avail = True
link = global_link
stem = "https://meta.wikimedia.org"
page_num = 1
global_rows = []
while (next_avail):
    if page_num % 100 == 0:
        print('Parsing page #{}'.format(page_num))
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    l = soup.find('ul', {"class":"mw-logevent-loglines"})
    items = l.find_all('li')
    for i in items:
        internal_links = i.find_all('a')
        try: 
            date = internal_links[0].text.strip()
            original = internal_links[4].text.strip()
            target = internal_links[5].text.strip()
            global_rows.append([date, original, target])
        except IndexError:
            print(i.text.strip())        
    next_link = soup.find('a', {"class":"mw-nextlink"})
    if next_link:
        link = stem + next_link['href']
        page_num += 1
    else:
        next_avail = False
        
```

```python
data_dir = "/Users/Lara/Documents/Stanford/Research/Network/processed_data/"
header = ["timestamp", "original", "target"]
global_df = pd.DataFrame(global_rows, columns=header)
global_df.to_csv(os.path.join(data_dir, 'global_renaming_log.csv'), index=False)
```
